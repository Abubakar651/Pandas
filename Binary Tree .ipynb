```python
class Node:
  def __init__(self,value):
    self.value=value
    self.left=None
    self.right=None

class BinarySearchTree:
  def __init__(self):
    self.root=None

  def insert(self,value):
    if self.root==None:
      self.root=Node(value)
    else:
      self._insert(self.root,value)

  def _insert(self,current_node,value):
    if value < current_node.value:
      if current_node.left==None:

        current_node.left=Node(value)
      else:

        self._insert(current_node.left,value)

    elif value > current_node.value:


      if current_node.right==None:
        current_node.right=Node(value)

      else:
        self._insert(current_node.right,value)

  def find(self,value):

    if self.root==None:

      return False

    else:
      return self._find(self.root, value)

  def _find(self,current_node,value):

    if current_node is None:
      return False

    if value==current_node.value:
      return True

    elif value < current_node.value:

      return self._find(current_node.left,value)

    else:
      return self._find(current_node.right,value)

  def inorder_traversal(self):

    result = []
    self._inorder_traversal(self.root, result)
    return result

  def _inorder_traversal(self, current_node, result_list):
    if current_node:
      self._inorder_traversal(current_node.left, result_list)

      result_list.append(current_node.value)

      self._inorder_traversal(current_node.right, result_list)


bst=BinarySearchTree()
bst.insert(1)
bst.insert(3)
bst.insert(6)
bst.insert(4)
bst.insert(5)
bst.insert(2)

print("inorder traversal: ",bst.inorder_traversal())
print("search 3: ", bst.find(3))
print("search 20: ", bst.find(20))
```

```python
class Stack:
  def __init__(self):
    self.stack = []

  def push(self,value):
    self.stack.append(value)

  def pop(self):
    if self.stack:
      self.stack.pop()
    else:
      print("Nothing in the stack it is empty")

  def empty(self):
    return len(self.stack)==0

st=Stack()
st.push(2)
st.push(3)
st.push(4)
st.push(5)
print("Stack",st.stack)
st.pop()
print("Stack after poping one element",st.stack)
print("Empty",st.empty())
```
